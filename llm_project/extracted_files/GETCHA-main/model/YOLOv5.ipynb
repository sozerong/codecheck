```python
import os

# 라벨 파일들이 위치한 경로 (Colab에서는 데이터를 적절한 경로로 설정)
label_folder = '/content/drive/MyDrive/colab/LSTM/yolov5/dataset/labels/train'  # 경로는 데이터에 맞게 수정

# 라벨 파일 내에서 클래스 번호를 수정하는 함수
def update_class_label(file_path, old_class, new_class):
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    with open(file_path, 'w', encoding='utf-8') as file:
        for line in lines:
            line_data = line.split()
            # 첫 번째 항목이 클래스 번호이므로, 해당 항목을 수정
            if line_data[0] == str(old_class):
                line_data[0] = str(new_class)
            file.write(' '.join(line_data) + '\n')

# 라벨 파일에 대해 클래스 번호 수정 적용 (예: 클래스 18을 클래스 1로 변경)
for label_file in os.listdir(label_folder):
    if label_file.endswith('.txt'):
        update_class_label(os.path.join(label_folder, label_file), old_class=18, new_class=1)
        update_class_label(os.path.join(label_folder, label_file), old_class=19, new_class=1)
```
